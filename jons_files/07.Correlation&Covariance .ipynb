```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
```

```python
def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n - 1)
```

```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)
```

```python
covariance(pageSpeeds, purchaseAmount), scatter(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount_1 = np.random.normal(50.0, 10.0, 1000) / pageSpeeds
plt.scatter(pageSpeeds, purchaseAmount_1)
```

```python
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x, y) / stddevx, stddevy

correlation(pageSpeeds, purchaseAmount)
```

```python
np.corrcoef(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = 100 - pageSpeeds * 3
scatter(purchaseAmount, pageSpeeds), correlation(purchaseAmount, pageSpeeds)
```

# Correlation Does NOT Imply Causality!!!
